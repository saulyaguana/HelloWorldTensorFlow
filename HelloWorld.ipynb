# Hello World in TensorFlow

+ For this example we will studying the correlation between the number of hours a student studied in a term and their final grade.
+ This is a theorical scenario, there are a lot more factors that come into play when it comes to how well a student will perform

We succesfully collated the hours of study of students and their corresponding grades as shown in the next table.

| Hours of study | 20 | 23 | 25 | 28 | 30 | 37 | 40 | 43 | 46 |
|----------------|----|----|----|----|----|----|----|----|----|
| Test Score     | 45 | 51 | 55 | 61 | 65 | 79 | 85 | 91 | 97 |

We want to build a model that predict how well a student will perform in the future based on the hours of study they put in.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
# Hours of study.
X = [20, 23, 25, 28, 30, 37, 40, 43, 46]

# Test Scores.
y = [45, 51, 55, 61, 65, 79, 85, 91, 97]
```

```python
# To get a good sense of data distribution, we use matplotlib to visualize our data
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(8, 6))
plt.plot(X, y)
plt.title("Exam Performance graph")
plt.xlabel("Hours of Study")
plt.ylabel("Test Score")
plt.show()
```

```python
# The plot above has a linear relationship.
# Logically when a student studies more hours he is gonna get a higher score.
# Without getting into a debate whether this theory holds, we are going to use the Keras API to build a simple model.
study_model = Sequential([Dense(units=1, input_shape=[1])])
study_model.compile(optimizer='adam', loss='mean_squared_error')

X = np.array(X, dtype=int)
y = np.array(y, dtype=int)
```

```python
# We build a one-layer model, which we call study_model.
# Now, we fit our model and run it for 2.500 epochs.
# fitting the model
history = study_model.fit(X, y, epochs=2500)
```
